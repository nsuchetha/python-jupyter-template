### Exercise 3: Scatter Plot with Annotations

```python
# TODO: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

# Data for the scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data Points')

# Annotate each point with its (x, y) value
for i, j in zip(x, y):
    plt.text(i, j, f'({i}, {j})', fontsize=12, ha='right', va='bottom')

# Add title and labels
plt.title('Scatter Plot with Annotations', fontsize=16)
plt.xlabel('X values', fontsize=14)
plt.ylabel('Y values', fontsize=14)

# Add legend
plt.legend(fontsize=12)

# Show the plot
plt.show()
```
