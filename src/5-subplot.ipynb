### Exercise 5: Subplot Layout

```python
# TODO: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

# Data for the plots
# Line chart
x_line = [0, 1, 2, 3, 4]
y_line = [0, 1, 4, 9, 16]

# Bar chart
categories_bar = ['A', 'B', 'C', 'D']
values_bar = [5, 7, 3, 9]

# Scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Pie chart
labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Line chart
axs[0, 0].plot(x_line, y_line, marker='o', color='blue', label='y = x^2')
axs[0, 0].set_title('Line Chart', fontsize=14)
axs[0, 0].set_xlabel('X', fontsize=12)
axs[0, 0].set_ylabel('Y', fontsize=12)
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot 2: Bar chart
axs[0, 1].bar(categories_bar, values_bar, color=['red', 'green', 'blue', 'purple'])
axs[0, 1].set_title('Bar Chart', fontsize=14)
axs[0, 1].set_xlabel('Categories', fontsize=12)
axs[0, 1].set_ylabel('Values', fontsize=12)

# Plot 3: Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='orange', label='Data Points')
for i, j in zip(x_scatter, y_scatter):
    axs[1, 0].text(i, j, f'({i}, {j})', fontsize=10, ha='right', va='bottom')
axs[1, 0].set_title('Scatter Plot', fontsize=14)
axs[1, 0].set_xlabel('X', fontsize=12)
axs[1, 0].set_ylabel('Y', fontsize=12)
axs[1, 0].legend()

# Plot 4: Pie chart
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart', fontsize=14)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
